**Netflix**! What started in 1997 as a DVD rental service has since exploded into one of the largest entertainment and media companies.

Given the large number of movies and series available on the platform, it is a perfect opportunity to flex your exploratory data analysis skills and dive into the entertainment industry. Our friend has also been brushing up on their Python skills and has taken a first crack at a CSV file containing Netflix data. They believe that the average duration of movies has been declining. Using your friends initial research, you'll delve into the Netflix data to see if you can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

You have been supplied with the dataset `netflix_data.csv` , along with the following table detailing the column names and descriptions:

## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Reading the Netflix data from the CSV file
netflix_df = pd.read_csv('netflix_data.csv')

# Displaying the first few rows of the Netflix dataframe
netflix_df.head()

# Creating a subset of the dataframe containing only movies
netflix_subset = netflix_df[netflix_df['type']=='Movie'] 

# Displaying the first few rows of the movie subset dataframe
netflix_subset.head()

# Creating a new dataframe with selected columns from the movie subset
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]

# Creating a new dataframe with movies that have a duration of 60 minutes or less
short_movies = netflix_movies[netflix_movies['duration'] <60]

# Creating a list of colors based on the genre of each movie
colors = []
for genre in netflix_movies['genre']:
    if genre == "Children":
        colors.append("red")
    elif genre == "Documentaries":
        colors.append("purple")
    elif genre == "Stand-up":
        colors.append("green")
    else:
        colors.append("blue")

# Creating a scatter plot of movie duration by year of release
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors, alpha=0.08)

plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')

# Creating a legend for the scatter plot
legend_labels = ['Children', 'Documentaries','Stand-up', 'Other']
legend_handles = [plt.scatter([],[], color=color, label=label) for color, label in zip(['red','purple', 'green', 'blue'], legend_labels)]
plt.legend(handles=legend_handles, title='Genre')

# Save the plot as an image file
plt.savefig('movie_duration_by_year.png')

# Show the plot
plt.show()

#Are we certain that movies are getting shorter? 
answer = "maybe"
print(answer)
```
